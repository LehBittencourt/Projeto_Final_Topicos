```python
import matplotlib.pyplot as plt
import numpy as np

def input_to_array(prompt):
    while True:
        try:
            return np.array([float(item) for item in input(prompt).split(',')])
        except ValueError:
            print("Por favor, insira apenas números separados por vírgulas.")

def plot_graph(x_values, y_values):
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, marker='o', color='b', linestyle='-', label='Dados do usuário')
    plt.title("Gráfico Bidimensional com Entradas do Usuário")
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.legend()
    plt.grid(True)
    plt.show()

x_values = input_to_array("Digite os valores de x separados por vírgulas (exemplo: 0,1,2,3,4,5):\n")
y_values = input_to_array("Digite os valores de y correspondentes, separados por vírgulas:\n")

plot_graph(x_values, y_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo CSV
csv_file = '/content/crescimento_populacional.csv'

# Tenta ler os dados do arquivo CSV
try:
    df = pd.read_csv(csv_file)
    if 'Ano' not in df.columns or 'População' not in df.columns:
        raise ValueError("CSV deve conter as colunas 'Ano' e 'População'")
except Exception as e:
    print(f"Erro ao ler o arquivo CSV: {e}")
    exit()

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 6))
plt.plot(df['Ano'], df['População'], marker='o', color='navy')
plt.title('Crescimento Populacional de 2010 a 2022 (Dados do CSV)')
plt.xlabel('Ano')
plt.ylabel('População (em bilhões)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Função para calcular z com base em x e y
def calculate_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Gerando valores de x e y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = calculate_z(x, y)

# Criando o gráfico de superfície 3D com mapa de cores 'viridis'
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
superficie = ax.plot_surface(x, y, z, cmap='viridis', rstride=1, cstride=1, antialiased=True)

# Adicionando barra de cores
fig.colorbar(superficie, shrink=0.5, aspect=10)

# Configurando rótulos dos eixos
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')

# Configurando o título
plt.title('Gráfico de Superfície com Mapa de Cores Viridis')

# Exibindo o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/temperaturas_2019.csv')
# Extracting x and y values from the dataframe
x = df['Mes'].values
y = df['Temperatura_Media'].values

# Creating a 2D grid of x and y
X, Y = np.meshgrid(x, y)

# Applying the function z = sin(x^2) + cos(y^2)
Z = np.sin(X**2) + np.cos(Y**2)

# Creating a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
surf = ax.plot_surface(X, Y, Z, cmap='plasma')

# Adding labels and a title
ax.set_xlabel('Mês')
ax.set_ylabel('Temperatura média (°C)')
ax.set_zlabel('Z = sin(x^2) + cos(y^2)')
ax.set_title('Gráfico de Superfície 3D para Dados de Temperatura com a Função Z = sin(x²) + cos(y²)')

# Adding a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```
